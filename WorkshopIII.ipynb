* สามารถสร้างแบบจำลองโดยใช้เทคนิค KNeighborsClassifier และข้อมูลทั้งหมด และทดสอบ
แบบจำลองโดยใช้ข้อมูลทั้งหมดได้ (Training set = testset) และแสดงข้อมูลผลลัพธ์ (y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้

```python
from sklearn import datasets
from sklearn import neighbors
from sklearn.metrics import confusion_matrix
from sklearn import metrics

#นำเข้าข้อมูล iris.data และ iris.target จาก sklearn
iris = datasets.load_iris()

#สร้าง Model KNeighborsClassifier
x = iris.data
print ("x = ", x ,"\n")
#แสดงผล Y_TRUE และ Y_Predict
y_true = iris.target
knn= neighbors.KNeighborsClassifier()
knn.fit(x, y_true)
y_pred = knn.predict(x)

print ("Y_TRUE = ", y_true,"\n")
print ("Y_Predict = ", y_pred,"\n")

#แสดงผล Confusion_matrix
print (confusion_matrix(y_true,y_pred))
print ("\n")

#แสดงผลการประเมินผลโมเดล ในส่วนของ Accuracy, Precision, Recall และ F-measure
print ("Accuracy = %.2f" %(metrics.accuracy_score(y_true,y_pred)))
print ("Precision = %.2f" %(metrics.precision_score(y_true,y_pred,average='weighted')))
print ("Recall = %.2f" %(metrics.recall_score(y_true,y_pred,average='weighted')))
print ("F-measure = %.2f" %(metrics.f1_score(y_true,y_pred,average='weighted')))
```

```python
from re import IGNORECASE
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score, KFold
import matplotlib.pyplot as plt

#นำเข้าข้อมูล iris จาก sklearn
iris = datasets.load_iris()

#สามารถสร้างแบบจำลองโดยใช้เทคนิค Artificial Neural Network และชุดข้อมูลที่แบง 8 ชุดนั้น (hint: MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000))
model = MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000)

#สามารถประเมินแบบจำลองโดยใช้ Accuracy ได้ โดยแสดงค่าของแต่ละรอบ และค่าเฉลี่ยทั้งหมด ออกทางจอภาพ
Accuracies = cross_val_score(estimator = model,X= iris.data,y= iris.target,cv= 8)
print(Accuracies)
Avg_Accuracy = Accuracies.mean()
print(Avg_Accuracy)


#สามารถแสดงกราฟเส้น (Line Graph) ของค่า Accuracy ทั้ง 8 ออกทางจอภาพ
plt.plot(Accuracies,color = '#692CFF')
plt.xlabel('Value of k of Knn')
plt.ylabel('Accuracy')
plt.show()
```

* สามารถนำเข้าข้อมูล iris data จาก sklearn datasets ได้ถูกต้อง และแบ่งข้อมูล Training set 80% และ Testset 20% โดยแสดงข้อมูล x_train, y_train, x_test, y_test ออกทางจอภาพได้ (x_train และ x_test ให้แสดงข้อมูลเพียงแค่ 10 rows เท่านั้น)
* สามารถสร้างแบบจำลองโดยใช้ 4 เทคนิค คือ KNeighborsClassifier, Decision Tree ความลึกสูงสุดของต้นไม้คือ 4, GaussianNB, และ Artificial Neural Network  โดยใช้ข้อมูล Training set และทดสอบแบบจำลองโดยใช้ข้อมูล testset และแสดงข้อมูลผลลัพธ์ (y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้
* สามารถแสดง Confusion Matrix ออกทางจอภาพได้
* สามารถประเมินแบบจำลองโดยใช้ F-measure ได้ โดยแสดงค่าทั้งหมดออกทางจอภาพ
* สามารถแสดงตารางเปรียบเทียบผลการประเมินของทุกเทคนิคได้

```python
from pandas.core.common import random_state
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import neighbors
from sklearn import tree
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
import pandas as pd

#สามารถนำเข้าข้อมูล iris data จาก sklearn datasets ได้ถูกต้อง และแบ่งข้อมูล Training set 80% และ Testset 20% 
iris = datasets.load_iris()
x = iris.data
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)


#โดยแสดงข้อมูล x_train, y_train, x_test, y_test ออกทางจอภาพได้
print("x_train = " ,x_train[0:10])
print("x_test = " ,x_test[0:10])
print("\n")

print("y_train = ",y_train)
print("y_test = " ,y_test)
print("\n")

knn = neighbors.KNeighborsClassifier()
knn.fit(x_test,y_test)
y_pred = knn.predict(x_test)
print("Y_TRUE (y_test) = %s" % y_test)

modelTree = tree.DecisionTreeClassifier(max_depth=4)
modelTree.fit(x_test,y_test)
y_pred2 = modelTree.predict(x_test)

modelG = GaussianNB()
modelG.fit(x_test,y_test)
y_pred3= modelG.predict(x_test)

modelMLP = MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000)
modelMLP.fit(x_test,y_test)
y_pred4= modelMLP.predict(x_test)

#แสดงข้อมูลผลลัพธ์ (y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้
print("Y_PREDICT (y_predict) (KN)= %s" % y_pred)
print("Y_PREDICT (y_predict) (Decision Tree)= %s" % y_pred2)
print("Y_PREDICT (y_predict) (GaussianNB)= %s" % y_pred3)
print("Y_PREDICT (y_predict) (Neural Network)= %s" % y_pred4)
print("\n")

#สามารถแสดง Confusion Matrix ออกทางจอภาพได้
print("Confusion_matrix (KN) = %s" % confusion_matrix(y_test, y_pred))
print("Confusion_matrix (Decision tree) = %s" %confusion_matrix(y_test, y_pred2))
print("Confusion_matrix (GaussianNB) = %s" %confusion_matrix(y_test, y_pred3))
print("Confusion_matrix (Neural Network) = %s" %confusion_matrix(y_test, y_pred4))
print("\n")

#สามารถประเมินแบบจำลองโดยใช้ F-measure ได้ โดยแสดงค่าทั้งหมดออกทางจอภาพ
print("F-measure (KN) = %.2f" % metrics.f1_score(y_test,y_pred,average='weighted'))
print("F-measure (Decision tree) = %.2f" % metrics.f1_score(y_test,y_pred2,average='weighted'))
print("F-measure (GaussianNB) = %.2f" % metrics.f1_score(y_test,y_pred3,average='weighted'))
print("F-measure (Neural Network) = %.2f" % metrics.f1_score(y_test,y_pred4,average='weighted'))
print("\n")

#สามารถแสดงตารางเปรียบเทียบผลการประเมินของทุกเทคนิคได้
F_KN = (metrics.f1_score(y_test,y_pred,average='weighted'))
F_DT = (metrics.f1_score(y_test,y_pred2,average='weighted'))
F_GNB = (metrics.f1_score(y_test,y_pred3,average='weighted'))
F_NN = (metrics.f1_score(y_test,y_pred3,average='weighted'))

F_Dic = [(F_KN,F_DT,F_GNB,F_NN)]

df = pd.DataFrame(F_Dic,columns=['KN','DecisionTree','GaussianNB','Neural Network'])
print(df)
```
